```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
admissions_df = pd.read_csv('admissions.csv',
    usecols=[
        'subject_id',
        'hadm_id',
        'admission_type',
        'admission_location',
        'discharge_location',
        'admittime',
        'dischtime',
        'insurance',
        'language',
        'marital_status',
         'race'

    ])

patients_df = pd.read_csv('patients.csv',
    usecols=[
        'subject_id',
        'gender',
        'anchor_age',
        'dod',


    ])

services_df = pd.read_csv('services.csv',
    usecols=[
        'subject_id',
        'hadm_id',
        'curr_service',


    ])
transfers_df = pd.read_csv('transfers.csv',
    usecols=[
        'subject_id',
        'hadm_id',
        'careunit',
        'intime',
        'outtime',
        'eventtype'


    ])
```

```python
# Création du dataset final enrichi
final_df = admissions_df[[
    'subject_id',
    'hadm_id',
    'admission_type',
    'admission_location',
    'discharge_location',
    'admittime',
    'dischtime',
    'insurance',
    'language',
    'marital_status',
    'race'
]]

# Ajout des features patients
patients_features = patients_df[[
    'subject_id',
    'gender',
    'anchor_age',
    'dod'
]]
final_df = final_df.merge(patients_features, on='subject_id', how='left')

# Ajout des features services
services_features = services_df[[
    'subject_id',
    'hadm_id',
    'curr_service',

]]
final_df = final_df.merge(services_features, on=['subject_id', 'hadm_id'], how='left')

# Ajout des features transfers
transfers_features = transfers_df[[
    'subject_id',
    'hadm_id',
    'careunit',
    'intime',
    'outtime',
    'eventtype'

]]
final_df = final_df.merge(transfers_features, on=['subject_id', 'hadm_id'], how='left')

# Vérification du résultat
print("\nFinal Dataset:")
print("Shape:", final_df.shape)
print("\nColonnes incluses:", final_df.columns.tolist())


final_df.to_csv('nanim.csv', index=False)
```
